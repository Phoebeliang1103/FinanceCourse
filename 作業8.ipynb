```python
import numpy as np
import scipy.stats as si
S = 90
K = 100
r = 0.01
sigma = 0.3
T_values = [1, 0.9, 0.5, 0.3, 0.1]

for T in T_values:
    d1 = (np.log(S/K) + (r + 0.5*sigma**2)*T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    call_price = S * si.norm.cdf(d1) - K * np.exp(-r*T) * si.norm.cdf(d2)
    print(f"T={T}, Call Price={round(call_price, 4)}")
```

```python
import numpy as np
import scipy.stats as si
import matplotlib.pyplot as plt

S = 90
K = 100
r = 0.01
sigma = 0.3
T_list = np.arange(1, 0.0, -0.1)
call_prices = []

for T in T_list:
    d1 = (np.log(S/K) + (r + 0.5*sigma**2)*T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    call_price = S * si.norm.cdf(d1) - K * np.exp(-r*T) * si.norm.cdf(d2)
    call_prices.append(call_price)

plt.figure(figsize=(8, 5))
plt.plot(T_list, call_prices, marker='o')
plt.title('European Call Option Price vs Time to Maturity')
plt.xlabel('Time to Maturity (T)')
plt.ylabel('Call Option Price')
plt.grid(True)
plt.gca().invert_xaxis()  # 讓 T 從 1 到 0.1 向左遞減
plt.show()
```
